```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gmaps

# from config import g_key
```

```python
#Read in each month csv

jan = pd.read_csv('Resources/202001-citibike-tripdata.csv')
feb = pd.read_csv('Resources/202002-citibike-tripdata.csv')
mar = pd.read_csv('Resources/202003-citibike-tripdata.csv')
apr = pd.read_csv('Resources/202004-citibike-tripdata.csv')
may = pd.read_csv('Resources/202005-citibike-tripdata.csv')
jun = pd.read_csv('Resources/202006-citibike-tripdata.csv')
```

```python
#For each month, append the next month, then drop all null values from the resulting data frame

jf = jan.append(feb)

jf.dropna()

jfm = jf.append(mar)

jfm.dropna()

jfma = jfm.append(apr)

jfma.dropna()

jfmam = jfma.append(may)

jfmam.dropna()

full_dataset = jfmam.append(jun)

full_dataset.dropna()

full_dataset.columns = ['tripduration', 'starttime', 'stoptime', 'start_stationid', 'start_stationname', 'start_stationlat', 'start_stationlon', 
             'end_stationid', 'end_stationname', 'end_stationlat', 'end_stationlon', 'bikeid', 'usertype', 'birthyear', 'gender']
```

```python
#Split the start and stop time columns in two, then place the date values into a new column
datetime = full_dataset["starttime"].str.split(" ", n = 1, expand = True) 

full_dataset['startdate'] = datetime[0]

enddatetime = full_dataset["stoptime"].str.split(" ", n = 1, expand = True) 

full_dataset['stopdate'] = enddatetime[0]
```

```python
#Get age by substrction birth year from 2020
full_dataset['age'] = 2020 - full_dataset['birthyear']
```

```python
#Get time in minutes by dividing time in seconds by 60
full_dataset['duration(min)'] = round(full_dataset['tripduration']/60, 2)

full_dataset
```

```python
del full_dataset['tripduration']
del full_dataset['starttime']
del full_dataset['stoptime']
del full_dataset['start_stationid']
del full_dataset['end_stationid']
del full_dataset['birthyear']

full_dataset
```

```python
clean_df = full_dataset[['duration(min)', 'startdate', 'stopdate', 'start_stationname', 'start_stationlat', 
             'start_stationlon', 'end_stationname', 'end_stationlat', 'end_stationlon', 
             'bikeid', 'usertype', 'age', 'gender' ]]

clean_df
```

```python
#Drop any row with age greater than 100
clean_dataframe = clean_df.loc[clean_df['age'] < 100, :]

clean_dataframe
```

```python
#Split clean df in two- one for subscribers, one for cusatomers

customers_df = clean_df.loc[clean_df['usertype'] == 'Customer', :]

subscribers_df = clean_df.loc[clean_df['usertype'] == 'Subscriber', :]
```

```python
#Get dates for each type
customer_dates = customers_df['startdate'].value_counts().sort_index()

subscriber_dates = subscribers_df['startdate'].value_counts().sort_index()

#Plot each ine separately
plt.plot(customer_dates, label="Customers")
plt.plot(subscriber_dates, label="Subscribers")

#Define month labels
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul']

#Plot xticks at the proper points with labels
plt.xticks(np.arange(0, 182, step=30), months, rotation=90)

plt.xlabel("Month")
plt.ylabel("Ridership")
plt.title("2020 NYC Citibike Ridership by Month")
plt.legend(loc='best')
```

```python
# scatter plot of stuff.
plt.scatter(clean_dataframe["age"],clean_dataframe["duration(min)"])
times=["10","20","30","40","50","60","70"]
tick_locations=[10000, 20000,30000,40000,50000,60000,70000]
plt.ylim(0,71000)
plt.yticks(tick_locations,labels=times)

plt.title("Age vs. Duration of Trip")
plt.ylabel("Duration (Minutes)")
plt.xlabel("Age")
```

```python
###Heatmap work

#Create grouped station dataframe
grouped_stations = pd.DataFrame(clean_dataframe['start_stationname'].value_counts().reset_index())

#Rename columns
mapping={grouped_stations.columns[0]: 'start_stationname',grouped_stations.columns[1]: 'trips'}

grouped_stations.rename(columns=mapping, inplace=True)

grouped_stations
```

```python
#Merge stations with clean dataframe
full_stations = pd.merge(grouped_stations, clean_dataframe, on='start_stationname', how='inner')

del full_stations['duration(min)']
del full_stations['startdate']
del full_stations['stopdate']
del full_stations['end_stationname']
del full_stations['end_stationlat']
del full_stations['end_stationlon']
del full_stations['bikeid']
del full_stations['usertype']
del full_stations['gender']
del full_stations['age']

full_stations = full_stations.drop_duplicates()

full_stations
```

```python
#Create heatmap layer


#Configure gmaps
gmaps.configure(api_key=g_key)

figure_layout = {
    'width': '700px',
    'height': '700px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}

fig = gmaps.figure(layout=figure_layout)

#Create heat layer
heat_layer = gmaps.heatmap_layer(full_stations[['start_stationlat', 'start_stationlon']], 
                                weights= full_stations['trips'], 
                                dissipating=False, max_intensity=15000,
                                point_radius=0.0015)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
# box plot for average trip duration per bike
duration_avg__bike_grouped = clean_dataframe.groupby('bikeid')['duration(min)'].mean()
# make dataframe
dur_avg_bike = pd.DataFrame(duration_avg__bike_grouped)

# Calculate the IQR and quantitatively determine if there are any potential outliers.  
quartiles = dur_avg_bike['duration(min)'].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
upperb = round(upperq + (1.5*iqr),2)
lowerb = round(lowerq - (1.5*iqr),2)
plt.boxplot(dur_avg_bike, sym = 'r')
plt.ylabel("Average duration (min)")
plt.title(f" What should go here?  Average bike trip duration across all bikes?")
plt.show()

## determine outliers using upper and lower bounds
# bike_dur_outliers = dur_avg.loc[(dur_avg["duration(min)"]<lowerq) | (dur_avg["duration(min)"]>upperq)]
# display(bike_dur_outliers)

# find max and min duration bike id
max_bike_dur = dur_avg_bike.max()
print(max_bike_dur)
min_bike_dur = dur_avg_bike.min()
print(min_bike_dur)
```
